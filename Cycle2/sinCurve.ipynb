```python
import numpy as np
import math
import matplotlib.pyplot as plt
from numpy.random import *
import pandas as pd
```

```python
# csv取り込み
df = pd.read_csv("./testSolve.csv", skiprows=1, names=('x', 'y'))
#ソート
df = df.sort_values('x')

print(df)
```

```python
#Xを作成
dfX = df.x
m = 3
X = np.ones((dfX.size, m + 1))

for i in range(dfX.size):
    for j in range(m):
        X[i, j + 1] = dfX[i] ** (j + 1)

#print(X)

#t
t = df.y

#wを求める
w = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), t)

#近似式
Z = np.zeros((2, dfX.size))
for i in range(dfX.size):
    Z[0, i] = dfX[i]
    for j in range(m+1):
        if j == 0:
            Z[1, i] = w[0]
        else:
            Z[1, i] = Z[1, i] + w[j]*(dfX[i]**j)

#Zをソートする必要がある


#図示
plt.scatter(df.x, df.y)
plt.plot(Z[0], Z[1], c='red')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```
